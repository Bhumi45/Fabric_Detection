**Affine Transformation**

```python
import matplotlib.pyplot as plt

img = cv.imread('E:\\Projects\\Fabric_Detection\\textures 3\\cotton\\46-scale_1_im_1_col.png')
assert img is not None, "file could not be read, check with os.path.exists()"
rows,cols,ch = img.shape
 
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
 
M = cv.getAffineTransform(pts1,pts2)
 
dst = cv.warpAffine(img,M,(cols,rows))
 
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
import cv2
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt


# Prepare dataset and labels
data = []
labels = []

# Define paths
dataset_dir = "E:\\Projects\\Fabric_Detection\\textures 3"
categories = ['cotton', 'corduroy', 'denim', 'linin', 'wool']

# Image Augmentation using TensorFlow
datagen = ImageDataGenerator(
    #rotation_range=15,        # Random rotations up to 15 degrees
  # width_shift_range=0.1,    # Horizontal shifts
  # height_shift_range=0.1,   # Vertical shifts
    horizontal_flip=True,     # Flip images horizontally
    vertical_flip=True,       # Flip images vertically
    #zoom_range=0.4,           # Random zoom
    brightness_range=[0.8, 1.2], # Brightness adjustment
    shear_range=0.4,
    #fill_mode="constant"           # Shear transformation
)

for category in categories:
    path = os.path.join(dataset_dir, category)
    #label = category
    
    for count,img_name in enumerate(os.listdir(path),start=1):

        if count==5:
            break

        img_path = os.path.join(path, img_name)
        image = cv2.imread(img_path)
        
        try:
            # Apply data augmentation and extract features
            image = cv2.resize(image, (128, 128))  # Resize to a fixed size
            for_aug_image = np.expand_dims(image, axis=0)  # Prepare for augmentation
            aug_iter = datagen.flow(for_aug_image, batch_size=1)

            # Perform 5 augmentations per image
            for _ in range(5):
                aug_img = next(aug_iter)[0].astype(np.uint8)
                

                plt.figure(figsize=(8, 4))

                #Original image
                plt.subplot(1, 2, 1)
                plt.imshow(image)
                plt.title(category)
                
                
                # Augmented image
                plt.subplot(1, 2, 2)
                plt.imshow(aug_img)
                plt.title("Augmented Image")
                

                plt.show()
        except Exception as e:
            print(e, img_path,img_name)
```
