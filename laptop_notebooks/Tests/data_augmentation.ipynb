```python
import cv2
import numpy as np
import os

# Define a function for augmentation (rotate, flip, etc.)
def augment_image(image):
    # Flip the image horizontally
    flipped_image = cv2.flip(image, 1)

    # Rotate the image by 45 degrees
    rows, cols, _ = image.shape
    M = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45, 1)
    rotated_image = cv2.warpAffine(image, M, (cols, rows))

    # Zoom in (scale up the image)
    zoomed_image = cv2.resize(image, None, fx=1.2, fy=1.2, interpolation=cv2.INTER_LINEAR)
    zoomed_image = zoomed_image[0:rows, 0:cols]  # Crop to the original size

    return flipped_image, rotated_image, zoomed_image

# Load an image using OpenCV
image_path = 'E:\\Projects\\Fabric_Detection\\textures 3\\cotton\\46a-scale_2_im_1_col.png'
image = cv2.imread(image_path)

# Directory to save augmented images
save_dir = 'Augmented_images'
os.makedirs(save_dir, exist_ok=True)

# Perform augmentations
flipped_image, rotated_image, zoomed_image = augment_image(image)

# Save the augmented images
cv2.imwrite(os.path.join(save_dir, 'flipped_image.jpg'), flipped_image)
cv2.imwrite(os.path.join(save_dir, 'rotated_image.jpg'), rotated_image)
cv2.imwrite(os.path.join(save_dir, 'zoomed_image.jpg'), zoomed_image)

print(f"Augmented images saved to {save_dir}")
```
